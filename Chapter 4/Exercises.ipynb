# Exercise 4.6

A particular **discrete** random variable, $X$, has the cdf

$$
F(x) = \left(\frac{x}{n}\right)^k; x = 1, 2, \ldots, n
$$

where $k$ and $n$ are constants characteristic of the underlying random phenomenon.  Determine $f(x)$, the pdf for this random variable, and, for the specific values $k=2, n=8$, compute and plot $f(x)$ and $F(x)$.

<hr>
<span style="color: #990033;">
   
### `Solution:`

$f(x) = F(x) - F(x-1)$

<br>
According to the given CDF,

$f(x) = \left(\frac{x}{n}\right)^k - \left(\frac{x-1}{n}\right)^k$

$\implies f(x) = \left(\frac{x^k - (x-1)^k}{n^k}\right) $

<br>
Substituting the values of $k$ and $n$ :

$f(x) = \left(\frac{x^2 - (x^2 + 1 - 2x)}{8^2}\right)$

$\implies f(x) = \left(\frac{2x-1}{64}\right); x = 1,2, \dots , 8$
<hr>

```python
def f(x):
    return((2*x - 1)/(64))

def F(x):
    return((x/8)**2)

x = list(range(1,9))
fx = np.array([f(i) for i in x])
Fx = np.array([F(i) for i in x])

plt.figure(figsize=(18, 5)) 

plt.subplot(1,3,1)
plt.bar(
    x,
    fx,
    label='f(x): p.d.f',
    width=0.9,
)
plt.legend()

plt.subplot(1,3,2)
plt.bar(
    x,
    Fx,
    label='F(x): c.d.f',
    width=0.9,
)
plt.legend()

plt.subplot(1,3,3)
k = [Fx[0]]
k.extend(list(Fx[1:] - Fx[:-1]))
plt.bar(
    x,
    k,
    label='F(x) - F(x-1) = pdf',
    width=0.9,
)
plt.legend()


plt.show()
```

# Exercise 4.7

The random variable, $X$, has the following pdf:

$$
f(x) = \left\{ \begin{array}{ll}
cx & 0<x<1 \\
0 & \mbox{otherwise}
\end{array}
\right.
$$

(i) First obtain the value of the constant, $c$, required for this to be a legitimate pdf, and then obtain an expression for the cdf $F(x)$.

<hr>
<span style="color: #990033;">
   
### `Solution:`

For this to be a legitimate pdf, $\int_0^\infty f(x)$ should be 1.

$
\int_0^\infty f(x) = \int_0^1 cx \, dx$

$
\implies c\int_0^1 x \, dx = 1
$

$\implies c \left. \frac{x^2}{2} \right|_0^1 = 1$

$\implies c \left[ \frac{1}{2} - \frac{0}{2} \right] = 1$

$\implies \frac{c}{2} = 1$

$\implies c = 2$

<br>
Now,

$f(x) = 2x$

$F(x) = \int_0^{x} f(x) \, dx$

$\implies F(x) = \int_0^{x} 2x \, dx= \left.x^2\right|_0^x$

<hr>

(ii) Obtain $P(X \le 1/2)$ and $P(X \ge 1/2)$.

<hr>
<span style="color: #990033;">
   
### `Solution:`

$P(X \le 1/2) = F(1/2) = \left.x^2\right|_0^{1/2}$

$\implies P(X \le 1/2) = \frac{1}{2}^2 = 0.25$

<br>

$P(X \ge 1/2) = F(X \le 1) - F(X \le 1/2)$

$\implies P(X \ge 1/2) = \left.x^2\right|_0^{1} - \left.x^2\right|_0^{1/2}$

$\implies P(X \ge 1/2) = \left.x^2\right|_{1/2}^{1}$

$\implies P(X \le 1/2) = 1 - \frac{1}{4} = 0.75$

> #### `Observations:`
> $P(X \ge x) = 1 - P(X \le x)$

<hr>

(iii) Obtain the value $x_m$ such that

$$
P(X \le x_m) = P(X \ge x_m)
$$

<hr>
<span style="color: #990033;">
   
### `Solution:`

$P(X \le x_m) = P(X \ge x_m)$

$\implies F(x_m) = 1 - F(x_m)$

$\implies F(x_m) = \frac{1}{2}$

$\implies \left.x^2\right|_0^{x_m} = \frac{1}{2}$

$\implies x_m^2 = \frac{1}{2}$

$\implies x_m = \frac{1}{\sqrt{2}}$

# Exercise 4.10

Show that $E(X)$ exists for the discrete random variable, $X$, with the pdf:

$$
f(x) = \frac{4}{x(x+1)(x+2)}; x = 1, 2, \ldots
$$

while $E(X)$ **does not** exist for the discrete random random variable with the pdf

$$
f(x) = \frac{1}{x(x+1)}; x = 1, 2, \ldots
$$

<hr>
<span style="color: #990033;">
   
### `Solution:`

The condition for existance of $E(X)$ is $\sum_1^\infty x \cdot f(x) < \infty$

<br>
For the first pdf:

$\sum_1^\infty x \cdot f(x) = \sum_1^\infty \frac{4x}{x(x+1)(x+2)}$

$= \sum_1^\infty \frac{4}{(x+1)(x+2)} $

$= 4\sum_1^\infty \frac{1}{(x+1)(x+2)}$

$= 4\sum_1^\infty \frac{(x+2) - (x+1)}{(x+1)(x+2)}$

$= 4\sum_1^\infty \left(\frac{(x+2)}{(x+1)(x+2)} - \frac{(x+1)}{(x+1)(x+2)}\right)$

$= 4\sum_1^\infty \left(\frac{1}{x+1} - \frac{1}{x+2}\right)$

$= 4\left[\left(\sum_1^\infty \frac{1}{x+1}\right) - \left(\sum_1^\infty \frac{1}{x+2}\right)\right]$

$= 4\left[\left(\frac{1}{1+1} + \frac{1}{2+1} + \frac{1}{3+1} + \dots \right) - \left(\frac{1}{1+2} + \frac{1}{2+2} + \frac{1}{3+2} + \dots \right)\right]$

$= 4\left[\left(\frac{1}{2} + \frac{1}{3} + \frac{1}{4} + \dots \right) - \left(\frac{1}{3} + \frac{1}{4} + \frac{1}{5} + \dots \right)\right]$

$= 4\left(\frac{1}{2} \right) = 2$

$\implies E(X)$ exists and is equal to 2.
